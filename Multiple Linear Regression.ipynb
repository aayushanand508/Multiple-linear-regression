# Dummy variables for the categorical variable state

```python

df["NewYork_State"]=np.where(df['State']=="New York", 1, 0)
df["California_State"]=np.where(df["State"]=="California", 1, 0)
df["Florida_State"]=np.where(df["State"]=="Florida", 1, 0)
```

```python
df.drop(columns=['State'],axis=1,inplace=True)
df.head()
```

```python
dependent_variable="Profit"
```

```python
#List of independent variables
independent_variables=df.columns.tolist()
```

```python
independent_variables.remove(dependent_variable)
```

```python
independent_variables
```

```python
#Data of independent variables
x=df[independent_variables].values

#Data of dpendent variables
y=df[dependent_variable].values
```

# Training Testing and Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
x_train[0:10]
```

# Applying Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
plt.scatter(y_test,y_pred)
plt.xlabel("Y_test")
plt.ylabel("Y_pred")
```

# Analysing the model

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
```

```python
sns.distplot((y_test,y_pred),bins=50)
```

Here, the curve is good so the our model is also good.

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

Again, the r2_score shows that our model is a good fit.
